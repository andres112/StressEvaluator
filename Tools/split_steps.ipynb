```python
import pandas as pd
import os

path = "./empatica/"
timestamp = "./timestamp_creator/output/timestamp.csv"
```

```python
timestamp_df = pd.read_csv(timestamp)
sessions= timestamp_df[['empaticaId','e4_diff_step_0','dur_step_0','e4_diff_step_2','dur_step_2','e4_diff_step_5','dur_step_5']]
sessions
```

```python
for i, session in sessions.iterrows():
    temp_folder = path+""+str(int(session['empaticaId']))
    temp_df = pd.read_csv(temp_folder+"/TEMP.csv")
    
    diff_base = int(4 * session['e4_diff_step_0']) + 1 # 1 is the offset in csv
    duration_base = int(4 * session['dur_step_0']) + 1
    baseline = temp_df.iloc[diff_base:diff_base + duration_base]
    baseline['markers'] = 0
    baseline.iloc[0:1]['markers'] = 1
    baseline.to_csv(temp_folder+"/TEMP_BASE.csv", index=False, header=None)
    
    diff_1 = int(4 * session['e4_diff_step_2']) + 1 # 1 is the offset in csv
    duration_1 = int(4 * session['dur_step_2']) + 1
    stressor_1 = temp_df.iloc[diff_1:diff_1 + duration_1]
    stressor_1['markers'] = 0
    stressor_1.iloc[0:1]['markers'] = 1
    stressor_1.to_csv(temp_folder+"/TEMP_S1.csv", index=False, header=None)
    
    
    diff_2 = int(4 * session['e4_diff_step_5']) + 1 # 1 is the offset in csv
    duration_2 = int(4 * session['dur_step_5']) + 1
    stressor_2 = temp_df.iloc[diff_2:diff_2 + duration_2]
    stressor_2['markers'] = 0
    stressor_2.iloc[0:1]['markers'] = 1
    stressor_2.to_csv(temp_folder+"/TEMP_S2.csv", index=False, header=None)
```

```python
for i, session in sessions.iterrows():
    temp_folder = path+""+str(int(session['empaticaId']))
    eda_df = pd.read_csv(temp_folder+"/EDA.csv")
    
    diff_base = int(4 * session['e4_diff_step_0']) + 2 # 2 is the offset in csv
    duration_base = int(4 * session['dur_step_0']) + 2
    baseline = eda_df.iloc[diff_base:diff_base + duration_base]
    baseline['markers'] = 0
    baseline.iloc[0:1]['markers'] = 1
    baseline.to_csv(temp_folder+"/EDA_BASE.csv", index=False, header=None)
    
    diff_1 = int(4 * session['e4_diff_step_2']) + 2 # 2 is the offset in csv
    duration_1 = int(4 * session['dur_step_2']) + 2
    stressor_1 = eda_df.iloc[diff_1:diff_1 + duration_1]
    stressor_1['markers'] = 0
    stressor_1.iloc[0:1]['markers'] = 1
    stressor_1.to_csv(temp_folder+"/EDA_S1.csv", index=False, header=None)
    
    diff_2 = int(4 * session['e4_diff_step_5']) + 2 # 2 is the offset in csv
    duration_2 = int(4 * session['dur_step_5']) + 2
    stressor_2 = eda_df.iloc[diff_2:diff_2 + duration_2]
    stressor_2['markers'] = 0
    stressor_2.iloc[0:1]['markers'] = 1
    stressor_2.to_csv(temp_folder+"/EDA_S2.csv", index=False, header=None)
```
